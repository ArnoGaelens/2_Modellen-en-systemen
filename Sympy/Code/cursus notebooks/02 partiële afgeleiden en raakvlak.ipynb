We willen het raakvlak $\alpha$ opstellen aan het oppervlak $\Sigma $ gegeven door het  voorschrift 
$$ \Sigma \leftrightarrow z = f(x,y)= x y e^{x} $$
in het punt met coördinaten $x_0$ = 1 en $y_0=2$.

De vergelijking van $\alpha $in een punt $P_0(x_0,y_0,z_0)$ is 
$$ z-z_0 = \left( \frac{ \partial{f} }{\partial{x}} \right)_0 (x-x_0) + \left( \frac{ \partial{f} }{\partial{y}}\right)_0(y-y_0) $$
 

N.B. We gebruiken de subscript 's' om aan te geven dat het om de symbolische variabelen gaat. 

```python
from sympy import exp, symbols,diff
x_s, y_s, z_s = symbols('x y z')
z_s = x_s*y_s*exp(x_s)
z_s
```

Nu kunnen we $z_0$ berekenen: 

```python
z_0 = z_s.subs({x_s:1,y_s:2})
z_0
```

Vervolgens berekenen we de partiële afgeleide $ \left( \frac{ \partial{f} }{\partial{x}}\right)_0  $ in $P(1,2)$:

```python
afg_x= diff(z_s,x_s)
afg_x.subs({x_s:1,y_s:2}) 
```

$\ldots$  en partiële afgeleide $ \left( \frac{ \partial{f} }{\partial{y}}\right)_0  $ in $P(1,2)$::

```python
afg_y= diff(z_s,y_s) 
afg_y.subs({x_s:1,y_s:2}) 
```

 Daarmee hebben we alle ingrediënten voor de vergelijking het raakvlak $\alpha$:  in $P(1,2)$
$$ z-2 \ e = 4 \ e (x- 1) +  e (y-2) $$
We vormen dit voorschrift om naar de functie $z= g(x,y)$:
$$ z  = g(x,y) =   4 \ e (x- 1) +  e (y-2) +2e $$
Wat neerkomt op: 
$$ z  = g(x,y) =   e  ( 4x+ y -4)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

lichtblauw = "#EEF4F7"
standaardblauw = "tab:blue"
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

opdeling = 100
x  = np.linspace(0,2,opdeling)
y  = np.linspace(0,3,opdeling)
z  = np.linspace(-3,40,opdeling)

X,Y = np.meshgrid(x ,y)
f = lambda x, y: x*y*np.exp(x)
Z_sigma = f(X, Y)
g = lambda x,y: exp(1)*(4*x+y-4)
Z_alpha = g(X,Y)


#plot van het vlak sigma
ax.plot_surface(X, Y, Z_sigma,cmap='viridis')
#plot van het vlak alpha
ax.plot_surface(X, Y, Z_alpha,color=lichtblauw)

#we tekenen een verticale lijn ter hoogte van het raakpunt
x_line = [1, 1]
y_line = [2, 2]
z_line = [0, 40]  

# Plot the vertical line
ax.plot(x_line, y_line, z_line, color='r', linestyle='-', lw=1)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.grid(False)
plt.show()
```
