```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
#from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection

lichtblauw = "#EEF4F7"
standaardblauw = "tab:blue"
```

```python
x=y=np.linspace(-3,3,100)
X,Y=np.meshgrid(x,y)
f = lambda x, y: np.sqrt(x**2+y**2)
Z = f(X, Y)
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax._axis3don = False
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.grid(False)
ax.plot_wireframe(X,Y,Z)
plt.show()
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

ax.contour(X,Y,Z) 
plt.show()
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

elev = 90 #elevation: hoek met grondvlak 
azim = 0 #Azimuth: hoek met x-as 
roll = 0 #roll: hoek waaronder het tafereel is gedraaid t.o.v. de verticale middenas 

ax.view_init(elev=elev, azim=azim, roll=roll)
ax.set_proj_type('ortho')

ax.contour(X,Y,Z)  
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

opdeling = 100
x  = np.linspace(0,2,opdeling)
y  = np.linspace(0,3,opdeling)
z  = np.linspace(-3,40,opdeling)

X,Y = np.meshgrid(x ,y)
f = lambda x, y: x*y*np.exp(x)
Z = f(X, Y)

#we maken een extra raster aan met waarden in het X, en Z-vlak, als voorbereiding op het plotten van het vlak alpha
X_vlak1, Z_vlak1 = np.meshgrid(x , z )


#plot van het vlak simga
ax.plot_surface(X, Y, Z,cmap='viridis',alpha=0.2)
    
#plot van vlak y = 2
ax.plot_surface(X_vlak1,  2 , Z_vlak1, alpha=0.8, color=lichtblauw)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.grid(False)
plt.show()
```

```python
# Genereer parameterwaarden t
t = np.linspace(0, 30, 100)

# stel de parametervergelijkingen op voor x,y en z
x = np.cos(t)
y = np.sin(t)
z = t

# Maak een 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot de ruimtekromme
ax.plot(x, y, z, label='Ruimtekromme')

# Voeg labels toe
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Toon de plot
plt.show()
```
