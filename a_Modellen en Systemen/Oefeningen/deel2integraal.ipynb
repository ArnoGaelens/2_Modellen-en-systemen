```python
from sympy import*
import sympy as sp
x, y = sp.symbols('x y')



```

```python
%matplotlib widget
from sympy.plotting import plot_implicit
vglelipse = sp.Eq((x/1)**2+(y/2)**2,1)
vglcirkel = sp.Eq((x**2+y**2),0.25)
p1 = plot_implicit(vglelipse, aspect_ratio=(1,1))
p2 = plot_implicit(vglcirkel, aspect_ratio=(1,1))

p1.extend(p2)

p1.show()
```
