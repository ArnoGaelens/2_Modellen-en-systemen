```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
#from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection

lichtblauw = "#EEF4F7"
standaardblauw = "tab:blue"
```

```python
x=y=np.linspace(-5,5,100)
#De 100 in linspace geeft aan dat tussen -5 en +5 de grafiek op 100 punten zal berekend worden. 
#Als blijkt (op de figuur) dat de resolutie te klein is (koekige lijnen ans je vloeiende verwacht) kan je voor dit getal 
#een grotere waarde kiezen, bvb. 1000 
X,Y=np.meshgrid(x,y)
f = lambda x, y: (100/(5+(x/3)**4 +y**2))
Z = f(X, Y)
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax._axis3don = True
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.grid(True)
ax.plot_wireframe(X,Y,Z)
plt.show()
```

```python
#Teken de niveaukrommen. Dit is standaard een 3D-figuur in dit softwarepakket!

fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

ax.contour(X,Y,Z,50) 
plt.show()
```

```python
#Omzetten van de 3D-contourplot naar een 2D-contourplot (klassieke niveaukrommenfiguur).
#Dit doe je hier met een trukje van de foor: je bekijkt de 3D-plot van bovenaf, evenwijdig met de z-as.

fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

elev = 90 #elevation: hoek met grondvlak 
azim = 270 #Azimuth: hoek met x-as 
roll = 0 #roll: hoek waaronder het tafereel is gedraaid t.o.v. de verticale middenas 

#Merk op dat azimuth=270graden om de x-as op de klassieke manier van links naar rechts te laten wijzen, en de y-as verticaal,
#je kijkt nu evenwijdig met de z-as van boven op de 3D-contourplot, je ziet dus een 2D-contourplot (niveaukrommen).  

ax.view_init(elev=elev, azim=azim, roll=roll)
ax.set_proj_type('ortho')

ax.contour(X,Y,Z,50)  
plt.show()
```

```python
#Om het berekende traject y=f(x) voor te stellen op een 3D-figuur moeten we een ruimtekromme gebruiken (zie later in de cursus).
#We krijgen hier een klassieke 2D-kromme y=f(x) op een 3D-figuur door z=0 te stellen.

#Opmerking: gebruik de helpfunctie! Bijvoorbeeld via de help "Using Matplotlib" kan je verschillende kleuren voor de
#grafieken definiÃ«ren. Dit kan je hÃ©Ã©l ingewikkeld maken maar eigenlijk heb je alleen de basiskleuren nodig... 

t = np.linspace(0,2,100)

x = t
y = np.e**(81/(4*(1.9)**2))*np.e**(-(81/4)*(1/t**2))
z = 0

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z,color='k')

plt.show()
```

```python
#Nu in bovenaanzicht.

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.view_init(elev=elev, azim=azim, roll=roll)
ax.set_proj_type('ortho')

ax.plot(x, y, z,color='k')

plt.show()
```

```python
#Niveaukrommenplot (concentratie geurstof en het traject van de mestkever op 1 figuur)

fig = plt.figure(facecolor=lichtblauw, figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

ax.view_init(elev=elev, azim=azim, roll=roll)
ax.set_proj_type('ortho')

ax.contour(X,Y,Z,50)  
ax.plot(x, y, z,color='k')

plt.show()

#Opmerking: pas eventueel de resolutie van de grafiek aan als de grafieken hoekerig geplot worden. 
```
