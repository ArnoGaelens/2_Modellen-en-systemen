We tekenen een raakvlak aan de grafiek van een functie van 2 veranderlijken, het raakvlak is evenwijdig met een gegeven vlak. 

```python
%matplotlib notebook
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

lichtblauw = "#EEF4F7"
standaardblauw = "tab:blue"
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

opdeling = 100
x  = np.linspace(-5,5,opdeling)
y  = np.linspace(-5,5,opdeling)
z  = np.linspace(0,10,opdeling)

X,Y = np.meshgrid(x ,y)
f = lambda x, y: 10-(x**2 +2*y**2)
Z_sigma = f(X, Y)
g = lambda x,y: 1-2*x-y
Z_alpha = g(X,Y)
h= lambda x,y: 89/8 -2*x-y
Z_raakvlak = h(X,Y)

ax.set_zlim(0, 10)
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

#plot van het oppervlak sigma
ax.plot_wireframe(X, Y, Z_sigma,cmap='viridis',alpha=0.5)
#ax.plot_wireframe(X, y, Z_sigma,color='b')
#plot van het vlak alpha
ax.plot_surface(X, Y, Z_alpha,color=lichtblauw)
#plot van het raakvlak
ax.plot_surface(X,Y,Z_raakvlak,color='r')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.grid(False)
plt.show()
```
