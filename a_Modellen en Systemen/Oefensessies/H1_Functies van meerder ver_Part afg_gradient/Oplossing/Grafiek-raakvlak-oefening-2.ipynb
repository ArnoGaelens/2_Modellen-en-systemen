We willen het raakvlak $\alpha$ opstellen aan het oppervlak $\Sigma $ gegeven door het  voorschrift 
$$ \Sigma \leftrightarrow z = f(x,y)= (sin(x^2 + y^2))^2 $$
in het punt met coördinaten $x_0$ = 1 en $y_0=1$.

De vergelijking van $\alpha $ in een punt $P_0(x_0,y_0,z_0)$ is 
$$ z-z_0 = \left( \frac{ \partial{f} }{\partial{x}} \right)_0 (x-x_0) + \left( \frac{ \partial{f} }{\partial{y}}\right)_0(y-y_0) $$
 

N.B. We gebruiken de subscript 's' om aan te geven dat het om de symbolische variabelen gaat. 

```python
from sympy import symbols,diff,sin,cos
x_s, y_s, z_s = symbols('x y z')
z_s = (sin((x_s)**2 +(y_s)**2))**2
z_s
```

Nu kunnen we $z_0$ berekenen: 

```python
z_0 = z_s.subs({x_s:1,y_s:1})
z_0
```

Vervolgens berekenen we de partiële afgeleide $ \left( \frac{ \partial{f} }{\partial{x}}\right)_0  $ in $P(1,1)$:

```python
afg_x= diff(z_s,x_s)
afg_x.subs({x_s:1,y_s:1}) 
```

$\ldots$  en partiële afgeleide $ \left( \frac{ \partial{f} }{\partial{y}}\right)_0  $ in $P(1,1)$::

```python
afg_y= diff(z_s,y_s) 
afg_y.subs({x_s:1,y_s:1}) 
```

 Daarmee hebben we alle ingrediënten voor de vergelijking het raakvlak $\alpha$:  in $P(1,1)$
$$ z-(sin(2))^2 = 4sin(2)cos(2) (x- 1) +  4sin(2)cos(2) (y-1) $$
We vormen dit voorschrift om naar de functie $z= g(x,y)$:
$$ z  = g(x,y) =  4sin(2)cos(2) (x- 1) +  4sin(2)cos(2) (y-1) +
(sin(2))^2  $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

lichtblauw = "#EEF4F7"
standaardblauw = "tab:blue"
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

opdeling = 100
x  = np.linspace(-2,2,opdeling)
y  = np.linspace(-2,2,opdeling)
z  = np.linspace(0,2,opdeling)

X,Y = np.meshgrid(x ,y)
f = lambda x, y: (np.sin(x*x + y*y))**2
Z_sigma = f(X, Y)
g = lambda x,y: 4*np.sin(2)*np.cos(2)*(x-1)+4*np.sin(2)*np.cos(2)*(y-1)+(np.sin(2))**2
Z_alpha = g(X,Y)


#plot van het vlak sigma
ax.plot_surface(X, Y, Z_sigma,cmap='viridis')
#plot van het vlak alpha
ax.plot_surface(X, Y, Z_alpha,color=lichtblauw)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.grid(False)
plt.show()
```
