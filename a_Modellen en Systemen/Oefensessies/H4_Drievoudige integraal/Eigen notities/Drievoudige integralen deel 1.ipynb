```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
x,y=np.meshgrid(x,y)

f = lambda x,y: 10-3*(x**2 +y**2)

z = f(x,y)

fig = plt.figure()

ax = fig.add_subplot(projection='3d')

ax.plot_surface(x,y,z)

ax.set_aspect('equal')

plt.show
```
