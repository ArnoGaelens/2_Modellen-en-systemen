```python
from sympy import symbols, sqrt, Derivative
import sympy as sp
x,y,z,w,u = symbols('x y z w u')

g = ((x*y)+2*z)-w*(x+y+z)-u*(x**2+y**2+z**2-24)
g
```

```python
afg_x = Derivative(g, x)
afg_x
```

```python
afg_x.doit()
```

```python
afg_y = Derivative(g, y)
afg_y.doit()
```

```python
afg_z = Derivative(g, z)
afg_z.doit()
```

```python
afg_w = Derivative(g, w)
afg_w.doit()
```

```python
afg_u = Derivative(g, u)
afg_u.doit()
```

```python
lijst = sp.solve((afg_w.doit(),afg_u.doit(), afg_x.doit(), afg_y.doit(), afg_z.doit()), dict= True)
lijst
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Definieer de functie van drie variabelen
def f(x, y, z):
    return x * y + 2 * z

# Genereer roosterpunten voor x, y, z
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)

# Genereer krommecordinaten voor x en y
x_curve, y_curve = np.meshgrid(x_vals, y_vals)
z_curve = -x_curve - y_curve  # Uitdrukking van z in termen van x en y

# Bereken de waarden van de functie op elk punt van het rooster
z_grid = f(x_curve, y_curve, 0)

# Plot de functie en de kromme
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x_curve, y_curve, z_grid, color='b', label='Functie f(x, y, z) = xy + 2z')
ax.plot_surface(x_curve, y_curve, z_curve, alpha=0.5, color='r', label='Kromme x + y + z = 0')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D-plot van de functie en kromme')
plt.legend()
plt.show()
```
