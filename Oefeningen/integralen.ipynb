```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Define the symbolic variable
x = sp.symbols('x')

# Define the functions
f = x**2
g = x + 2

# Generate x values
x_vals = np.linspace(-5, 5, 100)

# Evaluate the functions
y_vals_f = sp.lambdify(x, f)(x_vals)
y_vals_g = sp.lambdify(x, g)(x_vals)

# Plot the functions
plt.plot(x_vals, y_vals_f, label='f(x) = x^2')
plt.plot(x_vals, y_vals_g, label='g(x) = x + 2')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of the functions f(x) and g(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
opl = sp.solve(g- f, x)
opl
```

```python
integraalgx = sp.Integral(g,(x,opl[0],opl[1]))
integraalgx
```

```python
integraal = sp.Integral(1, (y, f,g))
integraal 
```

```python
integraaltot = sp.Integral(integraal,(x, 0, opl[1]))
integraaltot
```

```python
integraaltot.doit()
```
