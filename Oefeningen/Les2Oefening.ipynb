```python
from sympy import symbols, init_printing, exp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

init_printing()
x, y, t = symbols('x y t')

# matplotlib text rendering ipv tex
plt.rcParams['text.usetex'] = False
Blauw = "tab:blue"
lichtblauw = "#EEF4F7"

fig = plt.figure(facecolor=lichtblauw, figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')


# Grenzen
ondergrens_x = -3
bovengrens_x = 3
ondergrens_y = -3
bovengrens_y = 3
ondergrens_t = -3
bovengrens_t = 3

opdeling_x = 60
opdeling_y = 100
opdeling_t = 100

x_vals = np.linspace(ondergrens_x, bovengrens_x, opdeling_x)
y_vals = np.linspace(ondergrens_y, bovengrens_y, opdeling_y)

# Alleen t=0
t_val = 0

X, Y = np.meshgrid(x_vals, y_vals)
Z = (100*(1 + np.exp(-t_val)))**5 / (5 + X**2 + Y**2)

ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('T')
plt.show()
```
