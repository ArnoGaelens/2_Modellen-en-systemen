```python
from sympy import sqrt, symbols, init_printing
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

init_printing()
x, y = symbols('x y')

# matplotlib text rendering ipv tex
plt.rcParams['text.usetex'] = False
Blauw = "tab:blue"
lichtblauw = "#EEF4F7"

fig = plt.figure(facecolor=lichtblauw, figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')


# Grenzen
ondergrens_x = -3
bovengrens_x = 3
ondergrens_y = -3
bovengrens_y = 3

opdeling_x = 60
opdeling_y = 100
x = np.linspace(ondergrens_x, bovengrens_x, opdeling_x)
y = np.linspace(ondergrens_y, bovengrens_y, opdeling_y)

X, Y = np.meshgrid(x,y)

#=========================================================
"""def f(x, y):
    return np.sqrt(x**2 + y**2)
"""

f = lambda x, y: np.sqrt(x**2+y**2)

Z = f(X, Y)

ax.plot_surface(X,Y,Z)
plt.show()

```
